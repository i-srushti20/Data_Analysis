```python
import matplotlib.pyplot as plt
import numpy as np

# Group data by the "Fav genre" column
genre_counts = data['Fav genre'].value_counts()

# List of colors to be used in the pie chart
colors = ['#ff9999', '#66b3ff', '#99ff99', '#c2c2f0', '#ffcc99', '#ffb3e6', '#c1e1dc', '#ff6666', '#ff33cc']

# Create a pie chart
plt.figure(figsize = (10, 8))
ax = genre_counts.plot(kind = 'pie', autopct = '%1.1f%%', startangle = 140, colors = colors, 
                       pctdistance = 0.7,  labeldistance = 1.2,  textprops = {'color': 'white', 'fontsize': 12})  

# Add title without background color
plt.title('Distribution of Favorite Genres', color = 'white', fontsize = 16, fontweight = 'bold')

# Customize the background color to black (for the chart area)
plt.gca().set_facecolor('black')  
plt.gcf().patch.set_facecolor('black')  

# Remove the y-axis label (not needed for pie charts)
plt.ylabel('')

# Ensure labels do not overlap by adjusting layout
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import seaborn as sns

# Group the data by 'While working' and 'Hours per day'
heatmap_data = data.groupby(['While working', 'Hours per day']).size().unstack(fill_value = 0)

# Prepare data for the heatmap
heatmap_data = heatmap_data.fillna(0)

# Plotting the heatmap with custom colors and black background
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, cmap='BuPu', fmt='d', cbar_kws={'label': 'Count of Hours Worked'}, 
            linewidths=0.5, annot_kws={'color': 'gray'})

# Title and labels
plt.title('Heatmap of Hours per Day While Working', fontsize=16, fontweight='bold', color='white')
plt.xlabel('Hours per Day', fontsize=12, fontweight='bold', color='white')
plt.ylabel('While Working', fontsize=12, fontweight='bold', color='white')

# Set background color to dark black
plt.gca().set_facecolor('black')
plt.gcf().patch.set_facecolor('black')

# Adjust layout for better appearance
plt.tight_layout()
plt.show()
```

```python
# Group the data by 'Age' and 'Fav genre', and count occurrences
grouped_data = data.groupby(['Age', 'Fav genre']).size().unstack(fill_value=0)

# Plotting the Line Chart for Age and Fav genre
plt.figure(figsize = (12, 6))

# Plot for each genre across ages
for genre in grouped_data.columns:
    plt.plot(grouped_data.index, grouped_data[genre], label = genre, linewidth = 2)

# Set background color to dark Black
plt.gca().set_facecolor('black')

# Add titles and labels
plt.title('Music Genre Preferences by Age', fontsize = 16, fontweight = 'bold')
plt.xlabel('Age', fontsize = 12, fontweight = 'bold')
plt.ylabel('Count of Respondents', fontsize = 12, fontweight = 'bold')
plt.legend(title = 'Favorite Genre')

# Display the plot
plt.grid(axis = 'y', linestyle = '--', alpha = 0.5)
plt.tight_layout()
plt.show()
```

```python
# Count the number of occurrences in the 'Anxiety' column
anxiety_counts = data['Anxiety'].value_counts().sort_index()

# Plot a line chart
plt.figure(figsize=(8, 6))
plt.plot(anxiety_counts.index, anxiety_counts.values, marker='o', color='#fbefcc', linewidth=2, markersize=8)

# Set background color to dark black
plt.gca().set_facecolor('black')

# Customize the chart
plt.title('Distribution of Respondents by Anxiety', fontsize=16, fontweight='bold', color='black')
plt.xlabel('Anxiety Level', fontsize=12, color='black')
plt.ylabel('Count of Respondents', fontsize=12, color='black')
plt.xticks(rotation=45, fontsize=10, color='black')
plt.yticks(fontsize=10, color='black')
plt.grid(True, axis='y', linestyle='--', alpha=0.5)

# Show the chart
plt.tight_layout()
plt.show()
```

```python
# Limit the data to the first 5000 records
data_limited = data.head(5000)

# Count occurrences of each unique 'Depression' value in the limited dataset
depression_counts = data_limited['Depression'].value_counts()

# Generate a list of colors for each bar
colors = ['#FF1493', '#00FA9A', '#FFD700', '#FF6347', '#8A2BE2', '#7FFF00', '#00BFFF', '#FF4500', '#FF69B4']
# Plotting the horizontal bar chart with different colors for each bar
plt.figure(figsize = (12, 7))
bars = plt.barh(depression_counts.index, depression_counts, color = colors, edgecolor = 'white')

# Add labels and title
plt.xlabel('Number of Responses', fontsize = 12, fontweight = 'bold')
plt.ylabel('Depression Level', fontsize = 12, fontweight = 'bold')
plt.title('Distribution of Responses for Depression', fontsize = 16, fontweight = 'bold')

# Set background color to dark Black
plt.gca().set_facecolor('black')

# Show the grid for better readability
plt.grid(True, axis='x', linestyle='--', alpha=0.7)

# Display the bar chart
plt.tight_layout()
plt.show()
```

```python
# Count the occurrences of each primary streaming service
streaming_service_counts = data['Hours per day'].value_counts().sort_index()

# Scatter plot for count of responses vs. Primary streaming service
plt.figure(figsize = (12, 6))
plt.scatter(streaming_service_counts.index, streaming_service_counts.values, 
            color = '#ffef96', alpha = 1.0, edgecolor = '#FFC300')
# Adding labels and title
plt.title('Scatter Chart of Count of Responses vs. Hours per day', fontsize = 16, fontweight = 'bold')
plt.xlabel('Hours per day', fontsize = 12)
plt.ylabel('Count of Responses', fontsize = 12)
plt.grid(True, linestyle = '--', alpha = 0.5)

# Set background color to dark Black
plt.gca().set_facecolor('black')

# Display plot
plt.tight_layout()
plt.show()
```

```python
# Calculate the occurrences for each frequency column
columns_to_plot = ['Frequency [Classical]', 'Frequency [Metal]', 'Frequency [Lofi]', 'Frequency [Rap]', 'Frequency [K pop]']

# Prepare data for each frequency column
frequency_counts = {}
all_indices = set()

for column in columns_to_plot:
    counts = data[column].value_counts().sort_index()
    frequency_counts[column] = counts
    all_indices.update(counts.index)

# Reindex all counts to ensure they share the same index range
all_indices = sorted(all_indices)
for column in columns_to_plot:
    frequency_counts[column] = frequency_counts[column].reindex(all_indices, fill_value = 0)

# Plotting the Area Chart for all frequencies
plt.figure(figsize = (14, 8))

colors = ['#FF5733', '#33FF57', '#3380FF', '#FFC300', '#DA33FF']
labels = ['Classical', 'Metal', 'Lofi', 'Rap', 'K pop']

plt.gca().set_facecolor('#000000')  # Black background for the plot 

for i, column in enumerate(columns_to_plot):
    plt.fill_between(all_indices, frequency_counts[column].values, alpha = 0.5, label = f'Frequency [{labels[i]}]', color = colors[i])
    plt.plot(all_indices, frequency_counts[column].values, linewidth = 2, color = colors[i])

# Add titles and labels
plt.title('Area Chart: Responses by Frequency of Music Genres', fontsize = 16, fontweight = 'bold')
plt.xlabel('Frequency Levels', fontsize = 12, fontweight = 'bold')
plt.ylabel('Count of Responses', fontsize = 12, fontweight = 'bold')

plt.legend(fontsize = 12, title = "Music Genres")  # legend added

plt.grid(True, axis ='y', linestyle = '--', alpha = 0.5) # To Display the grid

plt.tight_layout() # Plot show 
plt.show()
```
